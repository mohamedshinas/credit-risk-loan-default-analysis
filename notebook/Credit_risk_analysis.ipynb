#### Remove High-Missing Columns (>70%)

These columns have over 70% missing values:

```python
cols_to_drop = [
    'verification_income_joint',
    'annual_income_joint',
    'debt_to_income_joint',
    'months_since_90d_late'
]

df.drop(columns=cols_to_drop, inplace=True)
```

Reason:

These variables contain insufficient information for reliable analysis.

#### Handle Moderate Missing Values

```python
df['emp_title'].fillna('Unknown', inplace=True)
```

```python
df['income_group'] = df['income_group'].cat.add_categories('Unknown')
df['income_group'].fillna('Unknown', inplace=True)
```

```python
df['emp_length'].fillna(df['emp_length'].median(), inplace=True)
```

```python
df['months_since_last_delinq'].fillna(
    df['months_since_last_delinq'].median(),
    inplace=True
)
```

```python
df['months_since_last_credit_inquiry'].fillna(
    df['months_since_last_credit_inquiry'].median(),
    inplace=True
)
```

```python
df['num_accounts_120d_past_due'].fillna(
    df['num_accounts_120d_past_due'].median(),
    inplace=True
)
```

```python
df['debt_to_income'].fillna(df['debt_to_income'].median(),inplace=True)
```

Results:

The dataset contained missing values in several columns. Variables with a large number of missing records were removed, while the remaining missing values were handled using appropriate techniques such as median imputation for numerical variables and "Unknown" labels for categorical variables. This improved the overall quality and reliability of the dataset for further analysis.

```python
df.isnull().sum()
print(df.info())
```

#### Correcting Data Types
* To ensure all columns are in the correct format for accurate analysis and calculations.

```python
df.dtypes
```

```python
cat_cols = [
    'state',
    'homeownership',
    'verified_income',
    'loan_purpose',
    'application_type',
    'grade',
    'sub_grade',
    'issue_month',
    'loan_status',
    'initial_listing_status',
    'disbursement_method'
]

for col in cat_cols:
    df[col] = df[col].astype('category')
```

Summary: 

The data types of all variables were examined during preprocessing. Numerical variables were already stored in appropriate formats (int64 and float64), while relevant categorical variables were converted from object to category data types to improve efficiency and support analysis. No major data type corrections were required.

####  Creating Derived Columns

To create new meaningful columns from existing data to improve analysis and insight.

```python
df['income_group'] = pd.cut(
    df['annual_income'],
    bins=[0,50000,100000,float('inf')],
    labels=['Low','Medium','High']
)
```

```python
df['loan_size_group'] = pd.cut(
    df['loan_amount'],
    bins=[0,10000,20000,float('inf')],
    labels=['Small','Medium','Large']
)
```

```python
df['dti_risk_group'] = pd.cut(
    df['debt_to_income'],
    bins=[0,10,20,float('inf')],
    labels=['Low Risk','Medium Risk','High Risk']
)
```

```python
df['employment_category'] = pd.cut(
    df['emp_length'],
    bins=[0,2,5,50],
    labels=['New','Experienced','Senior']
)
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.drop('employment_category', axis=1, inplace=True)
```

```python
df[['income_group',
    'loan_size_group',
    'dti_risk_group']].head()
```

Summary:

Three derived variables were created to support the analysis: Income Group (Low, Medium, High), Loan Size Group (Small, Medium, Large), and DTI Risk Group (Low Risk, Medium Risk, High Risk). These variables improved data segmentation and facilitated more effective analysis of borrower profiles and loan characteristics.

### 1. Univariate Analysis
#### 1.1 Distribution of Loan Amount

```python
plt.figure(figsize=(8,5))
sns.histplot(df['loan_amount'], bins=30, kde=True)
plt.title('Distribution of Loan Amount')
plt.tight_layout()
plt.show()
```

#### Interpretation

The distribution shows that most borrowers apply for small to medium-sized loans, while only a few borrowers request larger loan amounts. This indicates that the majority of customers prefer moderate borrowing, reducing overall exposure to high-value loans.

#### 1.2 Distribution of Annual Income

```python

plt.figure(figsize=(8,5))
sns.histplot(df['annual_income'], bins=30, kde=True)
plt.title('Distribution of Annual Income')
plt.tight_layout()
plt.show()
```

#### Interpretation

The annual income distribution reveals that most borrowers belong to low- and middle-income groups. A smaller number of high-income borrowers are present, creating a right-skewed distribution. This helps understand the financial profile of loan applicants.

#### 1.3 Distribution of Interest Rate

```python
plt.figure(figsize=(8,5))
sns.histplot(df['interest_rate'], bins=25, kde=True)
plt.title('Distribution of Interest Rate')
plt.tight_layout()
plt.show()
```

#### Interpretation

The interest rate distribution indicates that most loans are issued within a moderate interest rate range. Higher interest rates are less common and are generally associated with borrowers who pose a higher lending risk.

#### 1.4 Credit Grade Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(x='grade', data=df)
plt.title('Distribution of Credit Grades')
plt.tight_layout()
plt.show()
```

#### Interpretation

The credit grade distribution shows how borrowers are categorized based on their creditworthiness. A higher concentration in better grades suggests a stronger borrower portfolio, while lower grades indicate increased credit risk.

#### 1.5 Loan Status Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(y='loan_status', data=df)
plt.title('Loan Status Distribution')
plt.tight_layout()
plt.show()
```

#### Interpretation

The loan status analysis provides insights into loan performance. A larger number of current or fully paid loans indicates healthier portfolio performance, whereas a higher proportion of defaults may signal increased lending risk.

### 2. Bivariate Analysis
#### 2.1 Average Loan Amount by Income Group

```python
plt.figure(figsize=(8,5))
sns.barplot(x='income_group', y='loan_amount', data=df)
plt.title('Average Loan Amount by Income Group')
plt.tight_layout()
plt.show()
```

#### Interpretation

The analysis shows that loan amounts generally increase with borrower income levels. Higher-income borrowers tend to qualify for larger loans due to greater repayment capacity and financial stability.

#### 2.2 Interest Rate by Credit Grade

```python
plt.figure(figsize=(8,5))
sns.barplot(x='grade', y='interest_rate', data=df)
plt.title('Interest Rate by Credit Grade')
plt.tight_layout()
plt.show()
```

#### Interpretation

Interest rates vary across credit grades, with lower credit grades generally receiving higher interest rates. This reflects the lender’s risk-based pricing strategy, where riskier borrowers are charged higher borrowing costs.

#### 2.3 Loan Purpose Analysis

```python
plt.figure(figsize=(10,6))
sns.countplot(y='loan_purpose', data=df)
plt.title('Loan Purpose Distribution')
plt.tight_layout()
plt.show()
```

#### Interpretation

The loan purpose analysis identifies the most common reasons borrowers seek financing. Understanding these borrowing patterns helps lenders design targeted financial products and assess risk associated with different loan purposes.

#### 2.4 DTI Risk Group vs Interest Rate

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='dti_risk_group', y='interest_rate', data=df)
plt.title('DTI Risk Group vs Interest Rate')
plt.tight_layout()
plt.show()
```

#### Interpretation

Borrowers in higher debt-to-income risk groups tend to receive higher interest rates. This suggests that lenders consider existing debt obligations when evaluating borrower risk and determining loan pricing.

#### 2.5 Homeownership vs Annual Income

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='homeownership', y='annual_income', data=df)
plt.title('Homeownership vs Annual Income')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Interpretation

The comparison indicates differences in income levels across homeownership categories. Homeowners generally exhibit higher and more stable incomes, while renters may show greater variation in earnings.

### 3. Multivariate Analysis

#### 3.1 Income Group vs Grade

```python
plt.figure(figsize=(10,6))

sns.countplot(
    x='income_group',
    hue='grade',
    data=df
)

plt.title('Income Group by Credit Grade')
plt.tight_layout()
plt.show()
```

#### Interpretation

This analysis examines the relationship between borrower income and credit quality. Higher-income borrowers are more likely to have favorable credit grades, while lower-income groups may display a wider range of credit profiles.

#### 3.2 Loan Size Group vs DTI Risk Group

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='loan_size_group',
    hue='dti_risk_group',
    data=df
)

plt.title('Loan Size Group by DTI Risk Group')
plt.tight_layout()
plt.show()
```

#### Interpretation

The analysis helps determine whether larger loans are associated with higher financial risk. A greater concentration of large loans among high-risk borrowers may indicate potential exposure to future credit losses.

### 4: Correlation Analysis

```python
plt.figure(figsize=(12,8))

corr_matrix = df.select_dtypes(include=['int64','float64']).corr()

sns.heatmap(
    corr_matrix,
    cmap='coolwarm',
    annot=False
)

plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

#### Interpretation

Correlation analysis was conducted to examine the relationships between numerical variables such as loan amount, annual income, interest rate, debt-to-income ratio, and credit utilization. The heatmap helps identify variables that move together and provides insights into factors that may influence borrower behavior and lending decisions.
